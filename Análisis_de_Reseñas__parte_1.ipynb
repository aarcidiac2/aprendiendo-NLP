## 2) Preprocesamiento Inicial y Limpieza del Texto 游녧

**Objetivo: Realizar un preprocesamiento con Regex (Expresiones Regulares).**

**Regex: Es una secuencia de caracteres que define un patr칩n de b칰squeda**.

Se utiliza para ***describir, identificar y manipular*** cadenas de texto de manera eficiente.

Con expresiones regulares, puedes buscar coincidencias, extraer subcadenas, reemplazar partes de un texto o validar formatos de datos.

Nos vamos a quedar solo con la data de inter칠s

```python
filter_data = original_data[["review_body", "stars"]].copy()
```

```python
filter_data.head(3)
```

Evaluar valores nulos

```python
filter_data.isnull().sum()
```

Limpiar el texto

```python
import re
import string
```

```python
def clean(text):
    # Convertir a min칰sculas
    text = str(text).lower()

    # Eliminar textos entre corchetes (ej.: etiquetas)
    text = re.sub(r'\[.*?\]', '', text)

    # Eliminar URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Eliminar etiquetas HTML
    text = re.sub(r'<.*?>+', '', text)

    # Eliminar signos de puntuaci칩n
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)

    # Eliminar saltos de l칤nea
    text = re.sub(r'\n', ' ', text)

    # Eliminar palabras que contienen n칰meros
    text = re.sub(r'\w*\d\w*', '', text)

    # Eliminar emojis y caracteres especiales (no ASCII)
    text = re.sub(r'[^\x00-\x7F]+', '', text)

    # Eliminar espacios extras al inicio y final
    text = text.strip()

    return text
```

Aplicar la funci칩n de limpieza a la columna 'review_body'

```python
filter_data["clean_review"] = filter_data["review_body"].apply(clean)
```

## 4) Stemming y Lematizaci칩n para Rese침as 游닄
**Objetivo: Refinar el preprocesamiento aplicando eliminaci칩n de stopwords, stemming y lematizaci칩n para ver su efecto en las rese침as.**

**Stemming:** Consiste en recortar las palabras para reducirlas a su ra칤z (por ejemplo, "comprando" a "compr").

Sin embargo, este m칠todo puede producir resultados poco legibles y perder matices sem치nticos.
  
**Lematizaci칩n:** Reduce las palabras a su forma can칩nica (o lema), preservando mejor el significado (por ejemplo, "comprando" se convierte en "comprar").

**Eliminaci칩n de Stopwords:**
Se utiliza para eliminar palabras muy comunes que no aportan informaci칩n relevante (como 란l, 띿e, 랓).

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
```

Descargar recursos necesarios en espa침ol

```python
nltk.download('stopwords')
stopword_es = set(stopwords.words('spanish'))
```

```python
# Definir el stemmer para espa침ol
stemmer_es = SnowballStemmer("spanish")

# Seleccionar un ejemplo de rese침a limpia
sample_text = filter_data["clean_review"].iloc[1]
print("Texto de ejemplo:", sample_text)

# Tokenizaci칩n con nltk
tokens = word_tokenize(sample_text, language="spanish")
```

```python
print("Tokens:", tokens)
```

```python
# Eliminar stopwords
tokens_no_stop = [word for word in tokens if word not in stopword_es]
print("Tokens sin stopwords:", tokens_no_stop)

# Aplicar stemming
stemmed_tokens = [stemmer_es.stem(token) for token in tokens_no_stop]
print("Tokens despu칠s de stemming:", stemmed_tokens)

# Aplicar lematizaci칩n usando spaCy
doc_es = nlp_es(sample_text)
lemmatized_tokens = [token.lemma_ for token in doc_es if token.text.lower() not in stopword_es]
print("Tokens despu칠s de lematizaci칩n (sin stopwords):", lemmatized_tokens)
```

## 5) Visualizaci칩n y An치lisis Exploratorio del Texto 游

**Objetivo: Visualizar la frecuencia de t칠rminos mediante nubes de palabras.**

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

### 游늸 Rese침as Limpias

```python
# Generar nube de palabras con las rese침as limpias
text = " ".join(review for review in filter_data["clean_review"])
wordcloud = WordCloud(stopwords=STOPWORDS, background_color="white", width=800, height=400).generate(text)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Nube de Palabras - Rese침as Limpias")
plt.show()
```

### 游늸 Rese침as Limpias + Stopwords + Stemming

```python
def clean_with_stopwords_and_stemming(text):
    # Eliminar stopwords
    text = " ".join([word for word in text.split() if word not in stopword_es])
    # Aplicar stemming
    text = " ".join([stemmer_es.stem(word) for word in text.split()])
    # Eliminar espacios extras al inicio y final
    return text.strip()
```

```python
filter_data["clean_review_stemming"] = filter_data["clean_review"].apply(clean_with_stopwords_and_stemming)

# Mostrar ejemplos de texto limpio vrs texto limpio avanzado
filter_data[['clean_review', 'clean_review_stemming']].head(3)
```

```python
# Generar nube de palabras con las Rese침as Limpias + Stopwords + Stemming
text = " ".join(review for review in filter_data["clean_review_stemming"])
wordcloud = WordCloud(stopwords=STOPWORDS, background_color="white", width=800, height=400).generate(text)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Nube de Palabras - Rese침as Limpias + Stopwords + Stemming")
plt.show()
```

### 游늸 Rese침as Limpias + Stopwords + Lemmatization

```python
def clean_with_stopwords_and_lemmatization(text):
    # Procesar el texto usando spaCy
    doc = nlp_es(text)
    # Eliminar stopwords y aplicar lematizaci칩n
    lemmatized = [token.lemma_ for token in doc if token.text.lower() not in stopword_es]
    # Unir los tokens lematizados y eliminar espacios extra
    return " ".join(lemmatized).strip()
```

```python
filter_data["clean_review_lemmatization"] = filter_data["clean_review"].apply(clean_with_stopwords_and_lemmatization)

# Mostrar ejemplos de texto limpio vrs texto limpio avanzado
filter_data[['clean_review', 'clean_review_lemmatization']].head(3)
```

```python
# Generar nube de palabras con las Rese침as Limpias + Stopwords + Lemmatization
text = " ".join(review for review in filter_data["clean_review_lemmatization"])
wordcloud = WordCloud(stopwords=STOPWORDS, background_color="white", width=800, height=400).generate(text)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Nube de Palabras - Rese침as Limpias + Stopwords + Lemmatization")
plt.show()
```
